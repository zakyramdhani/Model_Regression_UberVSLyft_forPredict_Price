```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('model_scaler1.pkl', 'rb') as file_1:
  scaler = pickle.load(file_1)

with open('model_encoder1.pkl', 'rb') as file_2:
  ohe = pickle.load(file_2)

with open('model_lin_reg1.pkl', 'rb') as file_3:
  model = pickle.load(file_3)
```

```python
# membuat data dummy dengan 30 baris random
np.random.seed(1)
distance = np.random.randint(1,10,10)
cab_type = np.random.choice(['Uber', 'Lyft'])
name = np.random.choice(['UberX','Taxi','UberPool','Black SUV','WAV','UberXL','Black','Lyft XL','Lux Black',
                         'Lyft','Lux Black XL','Lux','Shared'],10)
surge_multiplier = np.random.choice(['1.00','1.25','1.50','1.75','2.00','2.50','3.00'],10)
```

```python
# Memberi nama kolom
inference = pd.DataFrame({'distance':distance,'cab_type':cab_type,'name':name,'surge_multiplier':surge_multiplier})
inference
```

```python
# Memisahkan data kategorik dan numerikal untuk di lakukan scaling dan encoding
data_num = pd.DataFrame(inference['distance'])
data_cat = pd.DataFrame(inference[['name', 'cab_type']])
```

```python
# Feature Scaling and Feature Encoding

inf_num_scaled = scaler.transform(data_num)
inf_cat_encoded = ohe.transform(data_cat)
inf_num_scaled = pd.DataFrame(inf_num_scaled)

names = { 0: 'distance'}

data_inf_num_scaled = inf_num_scaled.rename(columns = names)
data_inf_num_scaled
```

```python
inf_cat_encoded
```

```python
cat_ohe= ['name', 'cab_type']
feature_names= ohe.get_feature_names_out(input_features= cat_ohe)

X_train_ohe= pd.DataFrame(inf_cat_encoded, columns=feature_names)
X_test_ohe= pd.DataFrame(inf_cat_encoded, columns=feature_names)

X_train_ohe
```

```python
surge = pd.DataFrame(inference['surge_multiplier'])
surge
```

```python
inf_final = pd.concat([data_inf_num_scaled,X_train_ohe,surge],axis = 1)
inf_final
```

```python
# Predict Inference-Set

y_pred_inf = model.predict(inf_final)
y_pred_inf
```

```python
# Convert into DataFrame

y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['Price - Prediction'])
y_pred_inf_df
```

```python
tabel_inference = pd.concat([inference, y_pred_inf_df], axis=1)
tabel_inference
```

```python
# Get Intercept and Slope

intercept = model.intercept_
slope = model.coef_

print('Intercept : ', intercept)
print('Slope : ', slope)
```
